The goal of every data is to get insights. Getting insight from an unclean data is self destructive whereas a data without insight is as lifeless as a dead fish because what's the point of having such data in the first place. Although insight is the goal, there are other steps that precedes getting insights. This includes data analysis and visualizations for cleaned data obviously. To draw the insights from the data we have to use thorough data questioning, visualizations and also the technical know-how of such data. 

The archive_twitter_master.csv was the datasets I looked into. Parts of the columns are of categorical type while some are numeric, strings, datetime. i made sure i utilize the aspect that some of the columns are categorical by using the group_by method and this led me to get some insights from the data such as :
1. Puppo and doggo are the most occurred dog_category in the tweets
2. Top 10 tweets by favorite counts and retweet counts.

The above were visualized with pie_chart and bar graph.



Also we used line charts and scatter plots to get the trends of the dataset as we got insight that 
1. Tweets with high favorite counts tend to have high retweet counts.


2. Puppo and doggo are the most occurring dog_category in the tweets and this accounts for their high favorite and retweet counts.


We also observed that Labrador retriever, Chihuahua and Golden retriever are the most dog types predicted in all image prediction levels.
