```python
#condition
criteria = [(df1['p1_dog'] == True), (df1['p2_dog'] == True), (df1['p3_dog'] == True)]

#predicted breed
breed = [df1['p1'], df1['p2'],df1['p3']]

#confidence
conf = [df1['p1_conf'], df1['p2_conf'],df1['p3_conf']]

df1['breed'] = np.select(criteria, breed, default='none')

df1['conf'] = np.select(criteria, conf, default = 0)
```
